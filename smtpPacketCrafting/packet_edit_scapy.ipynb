```python
def printraw(pcaps):
    maxlen = 0
    line=0
    for packet in pcaps:
        line+=1
        try:
            print (packet["Raw"].load)
            maxlen = max(len(packet["Raw"].load), maxlen)
            print(line, len(packet["Raw"].load))
        except:
            continue
printraw(pcaps)
```

Creating multiple packets from the email text available with existing maximum length packet data. In this pcap has only one payload with the email data and creating multiple packets.

```python
import copy
endack = pcaps[19]["TCP"].ack
midPcaps = []
lastseq = pcaps[18]["TCP"].seq
lastack = pcaps[18]["TCP"].ack
flg=0
for i in range(int(len(mailpayload)/maxlen)+1):
    print("========= "+str(i)+" ========")
    print (bytes(mailpayload[i*maxlen:(i+1)*maxlen], "utf-8"))
    lastId += 1
    pcaps[19]["Raw"].load = bytes(mailpayload[i*maxlen:(i+1)*maxlen], "utf-8")
    pcaps[19]["IP"].len = len(pcaps[19]["IP"])
    pcaps[19]["IP"].id = lastId
    if flg==0:
        flg=1
        pcaps[19]["TCP"].seq = lastack
        lastseq+=1
        pcaps[19]["TCP"].ack = lastseq
    else:
        flg=0
        pcaps[19]["TCP"].seq = lastseq
        lastack+=1
        pcaps[19]["TCP"].ack = lastack
    packet = copy.copy(pcaps[19])
    midPcaps.append(packet)
midPcaps[2]["TCP"].ack = endack
```

```python
def changePayload(pcaps, pktNo, newPayload):
    oldSize = len(pcaps[pktNo])
    print (pktNo,"--------")
    # change payload and insert byte
    if (type(newPayload)==str):
        newPayload = "\r\n".join([ i.strip() for i in newPayload.split("\n")])
    elif (type(newPayload)==bytes):
        print ("payload in byte.... Processing ..... ")
    else:
        print ("The new payload should be string or byte format. You inserted "+ type(newPayload) + " .")
        return pcaps
#     print(newPayload)
    pcaps[pktNo]["Raw"].load = newPayload
#     print ("payload length : ",len(pcaps[pktNo]["Raw"].load))
    #change IP len value
    pcaps[pktNo]["IP"].len = len(pcaps[pktNo]["IP"])
    
    #change seq and ack values of the next packets
    pktSynSrt = str(pcaps[pktNo]["TCP"].seq)[:3]
    diffLen = len(pcaps[pktNo]) - oldSize
#     print ("payload length difference : ",diffLen)
    for pcap in pcaps[pktNo+1:]:
        if (str(pcap["TCP"].seq)[:3]==pktSynSrt):
            pcap["TCP"].seq += diffLen
        elif (str(pcap["TCP"].ack)[:3]==pktSynSrt):
            pcap["TCP"].ack += diffLen
    return pcaps
```
